```python
import cv2
import numpy as np
import datetime
```

```python
import tensorflow as tf 
model1 = tf.keras.models.load_model('C:/EXTRAS(FM)/VGG16-FB.h5')
model2 = tf.keras.models.load_model('C:/Users/salon/Desktop/EXTRAS(FM)/model_grey-012.model')
```

```python
body_classifier = cv2.CascadeClassifier('C:/Users/salon/Desktop/EXTRAS(FM)/HS.xml')
face_clsfr=cv2.CascadeClassifier('C:/FaceMask/training_1/haarcascade_frontalface_default.xml')
```

```python
source=cv2.VideoCapture(0)
```

```python
labels_dict1={0:'BACK',1:'FRONT'}
labels_dict2={0:'MASK',1:'NO MASK'}
color_dict={0:(0,255,0),1:(0,0,255)}
```

```python
count=0
people=4
```

```python
def confirm(l):
    for i in range(len(l)-1):
        if l[i]!=l[i+1]:
            return False
    return True
```

```python
enter_msg = 'Please Enter'
wait_msg = 'Kindly wait for your turn'
no_mask_msg = 'Kindly wear a mask! '
```

```python
import time
```

```python
while True:
    
        
        l1=[]
        l2=[]


        while True:
            
            
            
    
            ret,frame=source.read()
            dt = str(datetime.datetime.now())
            strcount = "Number of customers inside : " + str(count)
            cv2.putText(frame,dt, (10,15),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2)
            cv2.putText(frame,strcount, (10,450),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2)
            
            
    
            body =body_classifier.detectMultiScale(frame, 1.2, 3) 
        
            gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            faces=face_clsfr.detectMultiScale(gray,1.3,5)
    
            if type(body) is np.ndarray:
        
                for (a,b,c,d) in body:
            
            
                    cropped_body=frame[b:b+d,a:a+c]
                    cropped_body=cv2.resize(cropped_body,(224,224))
                    img_array=np.array(cropped_body)
                    im_array=np.expand_dims(img_array,axis=0)
                    result1=model1.predict(im_array)
                    #time.sleep(0.5)
                    label1=np.argmax(result1,axis=1)[0]
                    
                    if(len(l1)>3):
                        l1=[]
            
                    l1.append(label1)
                
                    
                    
                    print("l1 - ",l1)
            
                    if(label1==1):
                
                        for (x,y,w,h) in faces:
                    
                    
                            face_img=gray[y:y+w,x:x+w]
                            resized=cv2.resize(face_img,(224,224))
                            normalized=resized/255.0
                            reshaped=np.reshape(normalized,(1,224,224,1))
                            result2=model2.predict(reshaped)

                            label2=np.argmax(result2,axis=1)[0]
                            
                            if(len(l2)>3):
                                l2=[]
                    
                            l2.append(label2)
                            print("l2 - ",l2)
                            
            
                            #cv2.putText(frame, labels_dict2[label2], (10,100),cv2.FONT_HERSHEY_SIMPLEX,0.8,color_dict[label2],2)
                            cv2.rectangle(frame,(x,y),(x+w,y+h),color_dict[label2],2)

                #cv2.putText(frame, labels_dict1[label1], (10,150),cv2.FONT_HERSHEY_SIMPLEX,0.8,color_dict[label1],2)
     
            
    
            if len(l1)==4 :
            
                a= confirm(l1)
                
                print("a - ", a)
                if a==True:
                    if l1[0]==0 and count>0:
                        count = count -1
                        print("count = ", count)
                    l1=[]
                
            
            
            
            
            
            if len(l2)==4:
                b= confirm(l2)
                print("b - ",b)
                if b==True:
                    if l2[0]==0 :
                        if count<people:
                            cv2.putText(frame,enter_msg, (225,375),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2)
                            count = count +1
                            print("count = ", count)
                        else:
                            cv2.putText(frame,wait_msg, (175,375),cv2.FONT_HERSHEY_SIMPLEX,1,(51,255,255),2)
                    elif l2[0]==1:
            
                        cv2.putText(frame,no_mask_msg, (200,375),cv2.FONT_HERSHEY_SIMPLEX,1,(0,128,255),2)
                    l2=[]
            
        
            
            
            
            cv2.imshow('LIVE',frame)
            key=cv2.waitKey(1)
            if (key==27):
            
                break
            
            
            
        cv2.destroyAllWindows()
        source.release()   
        
        
        
        
```
